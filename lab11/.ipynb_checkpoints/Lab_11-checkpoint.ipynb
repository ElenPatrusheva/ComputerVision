```python
# all plots will be displayed directly below the code cell that produced it
%matplotlib inline
from matplotlib import pyplot as plt
from IPython.display import clear_output
# set inline plots size
plt.rcParams["figure.figsize"] = (16, 10) # (w, h)
# remove grid lines
import numpy as np
import time, cv2, math
from typing import AnyStr, Any, Callable
```

```python
# function to read and resize an image
def read_and_resize(filename: str, grayscale: bool = False, fx: float = 1.0, fy: float = 1.0):
    if grayscale:
      img_result = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    else:
      imgbgr = cv2.imread(filename, cv2.IMREAD_COLOR)
      # convert to rgb
      img_result = cv2.cvtColor(imgbgr, cv2.COLOR_BGR2RGB)
    # resize
    if fx != 1.0 and fy != 1.0:
      img_result = cv2.resize(img_result, None, fx=fx, fy=fy, interpolation = cv2.INTER_CUBIC)
    return img_result

    
def show_in_row(list_of_images: list, titles: list = None, disable_ticks: bool = False):
  count = len(list_of_images)
  for idx in range(count):
    subplot = plt.subplot(1, count, idx+1)
    if titles is not None:
      subplot.set_title(titles[idx])
      
    img = list_of_images[idx]
    cmap = 'gray' if (len(img.shape) == 2 or img.shape[2] == 1) else None
    subplot.imshow(img, cmap=cmap)
    if disable_ticks:
      plt.xticks([]), plt.yticks([])
  plt.show()
```

```python
img = read_and_resize("paper_test.jpg")
show_in_row([img])
```

### Find paper

```python
#TODO
# Find paper corners

def find_paper_corners(color_img: np.array) -> np.array:
  # return np.array with shape (4, 2)
  # order: [top_left, top_right, bot_right, bot_left] 
    gray = cv2.cvtColor(color_img, cv2.COLOR_RGB2GRAY)
    img_blur = cv2.medianBlur(gray,5).astype('uint8')
    

    ret2,tresh = cv2.threshold(img_blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    plt.imshow(tresh, cmap='gray')
    
    contours, _ = cv2.findContours(tresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    contour = max(contours, key = cv2.contourArea)
    epsilon = 0.01*cv2.arcLength(contour,True)
    approx = cv2.approxPolyDP(contour,epsilon,True)
    pts = approx[:, 0, :]
    
    # create the right order 
    # https://www.pyimagesearch.com/2016/03/21/ordering-coordinates-clockwise-with-python-and-opencv/
    s = pts.sum(axis=1)
    rect = [0,0,0,0]
    rect[0] = pts[np.argmin(s)]
    rect[2] = pts[np.argmax(s)]
    diff = np.diff(pts, axis=1)
    rect[1] = pts[np.argmin(diff)]
    rect[3] = pts[np.argmax(diff)]
        
    return np.array(rect)#np.array([[100, 100], [100, 200], [200, 200], [200, 100]])
corners = find_paper_corners(img)
if corners.shape == (4, 2):
  print("Detected corners: ",corners)
  img_copy = img.copy()
  cv2.polylines(img_copy, [corners], True, (255,0,0), 3)
  show_in_row([img_copy])
else:
  print("Shape must contain array of shape (4,2) representing image coordinates of paper corners")
```

```python
# TODO

def undistort_perspective(img: np.array, corners: np.array) -> np.array:
    h = 2 * int(((corners[0][0] - corners[3][0])**2 + (corners[0][1] - corners[3][1])**2)**0.5)
    w = 2 * int(((corners[0][0] - corners[1][0])**2 + (corners[0][1] - corners[1][1])**2)**0.5)
    dest_points = np.array([[0, 0], [w, 0], [w, h], [0, h]])
    
    homography, status = cv2.findHomography(corners, dest_points)
    res = cv2.warpPerspective(img, homography, (w, h))
    return res

paper = undistort_perspective(img, corners)
show_in_row([paper])
```

```python
img = paper[300:-400,100:-200].copy()
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_blur = cv2.medianBlur(gray,5).astype('uint8')
th3 = cv2.adaptiveThreshold(img_blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY_INV,11,2)
plt.imshow(th3, cmap = 'gray')
empty = np.zeros_like(gray)
lines = cv2.HoughLines(th3,1,np.pi/180,200)
minLineLength = 300
maxLineGap = 10
lines = cv2.HoughLinesP(th3,1,np.pi/180,200,minLineLength,maxLineGap)
xs, ys = [], []

for line in lines:
    [x1,y1,x2,y2] = line[0]
    ys.append(y1)
    ys.append(y2)
    xs.append(x1)
    xs.append(x2)
    empty = cv2.line(empty,(x1,y1),(x2,y2),255,5)
plt.imshow(empty, cmap='gray')
xs, ys = np.array(xs), np.array(ys)
```

```python
def find_grid(xs, img, n_peaks, is_vertical):
    image = img.copy()
    n, bins, patches = plt.hist(x=xs, bins=100, color='#0504aa',
                            alpha=0.7, rwidth=0.85)
    plt.grid(axis='y', alpha=0.75)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('My Very Own Histogram')
    plt.text(23, 45, r'$\mu=15, b=3$')
    maxfreq = n.max()
    # Set a clean upper y-axis limit.
    plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
    plt.show()

    n_bins = len(bins)
    delta = int(n_bins/(n_peaks-1)/2)
    peaks = []
    for i in range(n_peaks):

        values = n[int(max(n_bins/(n_peaks-1)*i-delta, 0)): min(int(n_bins/(n_peaks-1)*i+delta), n_bins)]
        peaks.append(np.argmax(values)+int(max(n_bins/(n_peaks-1)*i-delta, 0)))

    actual_peaks=[]
    for i in range(n_peaks):
        approx_peak = bins[peaks[i]]
        actual_peaks.append(np.mean(xs[abs(xs-approx_peak)<3]))
    print('Peaks of histogram are: ', actual_peaks)
    
    for a in actual_peaks:
        if is_vertical:
            image = cv2.line(image,(int(a),0),(int(a),1500),(255,255,0),4)
        else:
            image = cv2.line(image,(0,int(a)),(1500,int(a)),(255,0, 0),4)
    plt.imshow(image)
    plt.show()
    return actual_peaks, image
vertical_lines, img = find_grid(xs, img, 6, True)
horisontal_lines, img = find_grid(ys, img, 22, False)
```

```python
def has_check(block):    
    if np.var(block)>100:
        return True
    return False

def find_answers(img, hor_lines, vert_lines):
    answers = []
    blocks = []
    toLetter = {1:'A', 2:'B', 3:'C', 4:'D'}
    for hn in range(1, len(hor_lines)-1):
        lower = int(hor_lines[hn])+27
        upper = int(hor_lines[hn + 1])-15
        checks = []
        for vn in range(1, len(vert_lines)-1):
            left = int(vert_lines[vn])+20
            right = int(vert_lines[vn+1])-20
            block = img[lower:upper, left:right]
            if has_check(block):
                checks.append(toLetter[vn])
        answers.append(checks)
    return answers
```
